```python
data = {'url' : [],
        'text' : []
       }
count = 0

for i in range(len(df_input)):
    
    
    topic_url = df_input['URL'].loc[i]    
    data['url'].append(topic_url)

    response = requests.get(topic_url,headers=headers)
    page_contents = response.text
    doc = BeautifulSoup(page_contents, 'html.parser')
    topic_des_tags = doc.find_all('div', {'class' : 'td-post-content'})
    des = topic_des_tags[0].text
    heading = re.findall("[\n][\d]\).*[\n]", des)
    for i in heading:
        des = des.replace(i,"")
    
    des = des.replace('\n',"")
    data['text'].append(des)
    count +=1
    print(count,end = ",")
        

    
```

```python
from nltk.corpus import stopwords
import nltk

def remove_stopword(text):
    stopword=nltk.corpus.stopwords.words('english')
    stopword.remove('not')
    a=[w for w in nltk.word_tokenize(text) if w not in stopword]
    return ' '.join(a)
data_df['no_stopwords'] = data_df['text'].apply(remove_stopword)
```

```python
sentiment_df = {'neg': [], 'neu': [], 'pos':[] }
```

```python
sia = SentimentIntensityAnalyzer()
for i in range(len(data_df)):
    k = sia.polarity_scores(data_df['no_stopwords'].loc[i])
    for i in k:
        if (i != 'compound'):
            sentiment_df[i].append(k[i])
        
```

```python
data_df = pd.concat([data_df, pd.DataFrame(sentiment_df)], axis=1)
```

```python
from nltk.corpus import cmudict
d = cmudict.dict()
def nsyl(word):
  return [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]] 

syllable = {
        'total_syllable' : [],
        'complex_words' : [],
        'syllable_word_count': []
}

for i in range(len(data_df)):
    words = re.findall(r'\w*',data_df['text'].loc[i])
    while("" in word):
        words.remove("") 

    temp_complex = 0
    syllable_count = 0
    s_count = 0
    syllable_word_count = 0
    for w in words:

        try:
            tempcount = nsyl(w)[0]
            if tempcount > 2:
                temp_complex+=1
                syllable_count +=tempcount
            else:
                syllable_count +=tempcount
            syllable_word_count+=1
        except:
            continue
        
    syllable['total_syllable'].append(syllable_count)
    syllable['complex_words'].append(temp_complex)
    syllable['syllable_word_count'].append(syllable_word_count)
    
    temp_complex = 0
    syllable_count = 0
    s_count = 0
    syllable_word_count = 0
```
